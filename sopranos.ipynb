# The Sopranos 
An American crime drama television series created by *David Chase*. The series follows *Tony Soprano* (James Gandolfini), a New Jersey Mafia boss who has panic attacks. He reluctantly begins seeing psychiatrist *Dr. Jennifer Melfi* (Lorraine Bracco), who encourages him to open up about his difficulties balancing his family life with his criminal life. Other important characters include Tony's family, Mafia colleagues, and rivals, most notably his wife *Carmela* (Edie Falco), his nephew and protégé *Christopher Moltisanti* (Michael Imperioli), and his uncle *Corrado "Junior" Soprano* (Dominic Chianese).
