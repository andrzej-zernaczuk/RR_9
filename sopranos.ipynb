## Viewership

The Sopranos viewership generally rose from modest early Season 1 numbers to its highest levels around Seasons 3 and 4. Season 4 contains the peak episode, reaching over 13 million viewers. After that, viewership gradually declined through Seasons 5 and 6A, mostly staying below earlier highs. Season 6B was lower overall, but the finale produced a major spike, showing renewed audience interest at the end of the series.

## Changes in viewership

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load file in format: S01E01: 3.45
df = pd.read_csv(
    "viewership.csv",
    sep=":",
    names=["episode_id", "viewership"],
    engine="python"
)

df["episode_id"] = df["episode_id"].str.strip()
df["viewership"] = df["viewership"].astype(float)

# Parse season and episode number
parsed = df["episode_id"].str.extract(r"S(?P<season>\d{2}[AB]?)E(?P<episode>\d{2})")
df["season"] = parsed["season"]
df["episode"] = parsed["episode"].astype(int)

# Episode-to-episode change
df["viewership_change"] = df["viewership"].diff()

plt.figure(figsize=(16, 6))
plt.plot(df["episode_id"], df["viewership_change"], marker="o")
plt.axhline(0, color="black", linewidth=1)

plt.title("The Sopranos: E2E Change in Viewership")
plt.xlabel("Episode")
plt.ylabel("Change in viewership, millions")
plt.xticks(rotation=90)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

*The Sopranos* viewership was highly variable, with frequent increases and decreases between consecutive episodes. The largest positive change occurred around the transition into **S04E01**, where viewership increased by almost **4 million**. Another major increase happened at the end of the series, from **S06BE08 to S06BE09**, where viewership rose by nearly **3.9 million**.

The largest drops also occurred after major spikes. For example, after **S04E01**, viewership fell by a little over **2 million** in the next episode. Earlier in the series, there were also sharp declines, including a drop of more than **2 million** for **S02E02** and another drop of about **3 million** for **S03E03**. Later seasons show smaller episode-to-episode movements overall, especially through much of **Season 6A**, where changes mostly stayed close to zero.
